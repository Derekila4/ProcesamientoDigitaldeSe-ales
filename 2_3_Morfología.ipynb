# Ejercicio 1
● Dadas dos imágenes binarias A y B como se
muestran a continuación

● Realizar las siguientes operaciones
booleanas:

○ NOT(A)

○ A AND B

○ A OR B

○ A XOR B

○ [ NOT(A) ] AND B

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

imgA = cv2.imread("A.png")
imgB = cv2.imread("B.png")

NotA = cv2.bitwise_not(imgA)
AandB = cv2.bitwise_and(imgA, imgB)
AorB = cv2.bitwise_or(imgB, imgA)
AxorB = cv2.bitwise_xor(imgA, imgB)
NotAandB= cv2.bitwise_and(NotA, imgB)

plt.figure(figsize=(20, 10))

# Original
plt.subplot(1, 7, 1)
plt.imshow(imgA)
plt.title("A")
plt.axis("off")

plt.subplot(1, 7, 2)
plt.imshow(imgB)
plt.title("B")
plt.axis("off")

plt.subplot(1, 7, 3)
plt.imshow(NotA)
plt.title("Not A")
plt.axis("off")

plt.subplot(1, 7, 4)
plt.imshow(AandB)
plt.title("A and B")
plt.axis("off")


plt.subplot(1, 7, 5)
plt.imshow(AorB)
plt.title("A or B")
plt.axis("off")

plt.subplot(1, 7, 6)
plt.imshow(AxorB)
plt.title("A xor B")
plt.axis("off")

plt.subplot(1, 7, 7)
plt.imshow(NotAandB)
plt.title("Not A and B")
plt.axis("off")


```

# Ejercicio 2
● Realiza un programa en Python que permita realizar la
dilatación y la erosión de una imagen donuts.png

● Intenta diferentes tamaños del elemento estructural B
que permita la mejor solución para rellenar los huecos
de los círculos

```python
img = cv2.imread("donuts.png",0)

kernel  = np.ones((15,15), np.uint8)

img_eorsion = cv2.erode(img, kernel, iterations= 1)

img_dilatacion= cv2.dilate(img, kernel, iterations=1)

cv2_imshow(img_eorsion)
cv2_imshow(img_dilatacion)
```

# Ejercicio 4
● Mediante las siguientes equivalencias, realiza un programa
en Python para calcular las operaciones de Opening y
Closing en la imagen fingerprint.png

● Puedes usar diferentes tamaños del elemento estructural B


```python
img = cv2.imread("fingerprint.png",0)

kernel  = np.ones((3,3), np.uint8)

kernelon= cv2.getStructuringElement(cv2.MORPH_ERODE, (3,3))
kernelon2= cv2.getStructuringElement(cv2.MORPH_DILATE, (3,3))

img_eorsion = cv2.erode(img, kernelon, iterations= 1)

img_dilatacion= cv2.dilate(img, kernelon2, iterations=1)

img_opening= cv2.dilate(img_eorsion, kernelon2, iterations=1)

img_closing = cv2.erode(img_dilatacion, kernelon, iterations=1)


cv2_imshow(img)
cv2_imshow(img_opening)
cv2_imshow(img_closing)
```

# Ejercicio 5

De la imagen fingerprint.png realiza las siguientes operaciones morfológicas:

● im1 = (A⚬B) ⊕ B

● im2 = [ (A⚬B) ⊕ B ] ⊖ B = (A⚬B)⚫B

```python
img = cv2.imread("fingerprint.png",0)

kernel  = np.ones((3,3), np.uint8)

kernelon= cv2.getStructuringElement(cv2.MORPH_ERODE, (3,3))
kernelon2= cv2.getStructuringElement(cv2.MORPH_DILATE, (3,3))

kernelon3= cv2.getStructuringElement(cv2.MORPH_ERODE, (4,4))

img_eorsion = cv2.erode(img, kernelon, iterations= 1)

img_dilatacion= cv2.dilate(img, kernelon2, iterations=1)

img_opening= cv2.dilate(img_eorsion, kernelon2, iterations=1)

img_closing = cv2.erode(img_dilatacion, kernelon, iterations=1)


im1= cv2.dilate(img_opening, kernelon3, iterations=1)

im2= cv2.erode(im1, kernelon, iterations=1)

cv2_imshow(img)
cv2_imshow(img_opening)
cv2_imshow(img_closing)
cv2_imshow(im1)
cv2_imshow(im2)
```

# Ejercicio 6

● De la imagen Lincoln_from_penny.png realiza la operación de extracción de
bordes mediante la ecuación:

𝜷(A) = A - ( A⊖B )

```python
import cv2
import numpy as np

img = cv2.imread("Lincoln_from_penny.png", 0)

kernel = np.ones((3, 3), np.uint8)

erosion = cv2.erode(img, kernel, iterations=1)

beta = img - erosion


cv2_imshow(beta)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
