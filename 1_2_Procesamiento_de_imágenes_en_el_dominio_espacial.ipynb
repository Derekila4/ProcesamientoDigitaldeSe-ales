# Ejercicio 2
● Realiza un programa en python que aplique la
corrección gamma de la imagen aérea que se
muestra en la imagen original mediante la
ecuación

s = c ( r + ε )𝛾

● En la figura se muestran los resultados de la
aplicación de la transformación con c = 1 para
diferentes valores de gamma

● Verifica qué sucede con valores de gamma
iguales a 2, 3, 4, 5, …

● Verifica qué sucede con valores de gamma
iguales a ½ , ⅓ , ¼ , ⅕ , …




```python
import matplotlib.pyplot as plt

img = cv2.imread('aerial.png',1)

e = 0.001
c = 1
gamma = 3

gamma3=c*(img+e)**gamma

gamma = 4

gamma4=c*(img+e)**gamma


gamma = 15

gamma5=c*(img+e)**gamma

gamma3 = cv2.normalize(gamma3, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)
gamma4 = cv2.normalize(gamma4, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)
gamma5 = cv2.normalize(gamma5, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)

fig, axs = plt.subplots(2, 2)

axs[0, 0].imshow(img)
axs[0, 0].set_xticks([]), axs[0, 0].set_yticks([])

axs[0, 1].imshow(gamma3)
axs[0, 1].set_xticks([]), axs[0, 1].set_yticks([])

axs[1, 0].imshow(gamma4)
axs[1, 0].set_xticks([]), axs[1, 0].set_yticks([])

axs[1, 1].imshow(gamma5)
axs[1, 1].set_xticks([]), axs[1, 1].set_yticks([])

plt.show()
```

# Ejercicio 3

Realizar un programa en Python que imprima el histograma de una imagen,
como se muestra a continuación

```python
img = cv2.imread('rose.png',0)

cv2_imshow(img)
ancho , largo= img.shape

x= np.linspace(0,255,num=256, dtype=np.uint8 )
y = np.zeros(256)

for a in range(ancho):
  for l in range (largo):
    v= img[a,l]
    y[v]=y[v]+1

plt.bar(x,y)
plt.show()

```

# Ejercicio 4

● Realizar un programa en Python que imprima el histograma de una imagen a
color, por cada canal RGB por separado, tal como se muestra a continuación.

● Verificar que canal de color predomina más y en qué valores de identidad se
encuentran


```python
img = cv2.imread('rose.png')
cv2_imshow(img)

canales = cv2.split(img)

plt.figure()
plt.xlim((0, 256))

colores = ('b', 'g', 'r')

for (canales, colores)in zip(canales, colores):
  histograma= cv2.calcHist([canales],[0],None,[256],[0,256] )
  plt.plot(histograma, color=colores)

plt.show()
```
