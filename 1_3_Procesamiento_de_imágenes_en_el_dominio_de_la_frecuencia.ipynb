# Ejercicio 1

● Desarrolla un programa en Python que
realice la transformada de Fourier de una
imagen.

● Crear una imagen de 512x512 píxeles de
fondo negro con un rectángulo blanco de
20x40 píxeles centrado en el origen, como la
que se muestra a la izquierda

● Calcular la transformada discreta de Fourier
y aplicar la operación logarítmica para
visualizarla con matplotlib, como se muestra
a la derecha

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.zeros((512,512))
img[256-10:256+10,256-20:256+20] = 1
img = np.float32(img)

plt.imshow(img, cmap='gray')

dft = cv2.dft (img, flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitud = np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:,:,1]))

magnitud = abs(magnitud**2)
plt.imshow(magnitud, cmap = 'gray')
plt.show()
```

# Ejemplo 3
● Realiza un Programa en Python que permita
visualizar la transformada de Fourier de:

○ Una imagen en escala de grises

○ Una imagen a color



```python
from scipy import fftpack
import matplotlib.pyplot as plt
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('peppersBW.png', 0)
img_float32 = np.float32(img)  # Convertir la imagen a tipo float32

plt.imshow(img_float32, cmap='gray')

dft = cv2.dft(img_float32, flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitud = np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer subgráfico
plt.imshow(img, cmap='gray')

# Mostrar la magnitud de la Transformada de Fourier
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo subgráfico
plt.imshow(magnitud, cmap='gray')
plt.show()

#IMAGEN A COLOR


# Leer la imagen en color
img = cv2.imread('peppers.png')

# Convertir la imagen a tipo float32
img_float32 = np.float32(img)

# Separar los canales de color
b, g, r = cv2.split(img_float32)

# Aplicar la transformada de Fourier a cada canal por separado
dft_b = cv2.dft(b, flags=cv2.DFT_COMPLEX_OUTPUT)
dft_g = cv2.dft(g, flags=cv2.DFT_COMPLEX_OUTPUT)
dft_r = cv2.dft(r, flags=cv2.DFT_COMPLEX_OUTPUT)

# Calcular la magnitud de la Transformada de Fourier para cada canal
magnitude_b = np.log(cv2.magnitude(np.fft.fftshift(dft_b[:, :, 0]), np.fft.fftshift(dft_b[:, :, 1])))
magnitude_g = np.log(cv2.magnitude(np.fft.fftshift(dft_g[:, :, 0]), np.fft.fftshift(dft_g[:, :, 1])))
magnitude_r = np.log(cv2.magnitude(np.fft.fftshift(dft_r[:, :, 0]), np.fft.fftshift(dft_r[:, :, 1])))

# Combinar los resultados de los canales
magnitude_combined = np.maximum(np.maximum(magnitude_b, magnitude_g), magnitude_r)

# Mostrar la imagen original y la magnitud de la Transformada de Fourier
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))


plt.subplot(1, 2, 2)
plt.imshow(magnitude_combined, cmap='gray')


plt.show()
```

# Ejemplo 5

● Del ejemplo anterior, programa en Python tu propia función de convolución y
compara la imagen de salida resultante con una medida de distancia (ejemplo
Euclideana)

```python

image = cv2.imread('cactus.png', 0)
image = image.astype(np.float32)

k = 6
kernel = np.ones((k, k), np.float32) / (k ** 2)


resultado = cv2.filter2D(image, -1, kernel)


I1 = image.astype(np.float32)
I2 = resultado.astype(np.float32)
DE = np.sqrt(np.sum((I1-I2)**2))
print("Distancia euclidiana entre las imágenes:", DE)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')


plt.subplot(1, 2, 2)
plt.imshow(resultado, cmap='gray')


plt.show()
```

# Ejemplo 6
● Del Ejemplo 4 usando g(x,y) = f(x,y) ✳ h(x,y) ahora compara el resultado utilizando la Convolución en el dominio de
la frecuencia mediante la transformada de Fourier de las dos señales

● Sea el par transformado: f(x,y) ✳ h(x,y) ⇔ F(u,v) H(u,v) entonces despejando, la imagen de salida se calcula con la
transformada inversa de Fourier del producto de las señales transformadas g(x,y) =IFT{ F(u,v) H(u,v) }

● Utiliza las funcion fft2() e ifft2() para calcular la transformada y transformada inversa de Fourier

```python
k=7

kernel = np.zeros(image.shape)

row, col= image.shape

row0,col0 = int(row/2)-int(k/2), int(col/2)-int(k/2)

row1, col1 = int(row0+k), int(col0+k)

kernel[row0:row1,col0:col1]= np.ones((k,k),np.float32)/k**2


F = np.fft.fft2(image)
H = np.fft.fft2(kernel)

Corr = F*H

Blur = np.fft.ifft2(Corr)
Blur = np.fft.fftshift(Blur)
Blur = abs(Blur)

cv2_imshow(Blur)
```

# Ejemplo 7

● Realiza un programa en python con la función correlate2D() y el paquete signal de scipy que aplique
la correlación cruzada de dos imágenes, donde h(x,y), es una región de interés de la imagen original f(x,y)

● Graficar la función de correlación usando plot() y contourf()

● Calcular la coordenada del máximo valor de correlación. Utilizar dicha coordenada en la imagen para
demostrar la detección de h(x,y) en la entrada f(x,y)

```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt



f = cv2.imread("ABC.png", 0)/255.
h = cv2.imread("A.png", 0)/255.
f -= np.mean(f)
h -= np.mean(h)

corr = signal.correlate2d(f , h, boundary ='symm', mode = 'same')
y, x = np.unravel_index(np.argmax(corr), corr.shape)

fig, axs = plt.subplots(nrows=1, ncols=2,figsize=(12, 5))

plt.subplot(121), plt.imshow(f, cmap ='gray')
plt.plot(x, y,'r*', markersize='10')


plt.subplot(122), plt.imshow(h, cmap = 'gray')

plt.show()

plt.contourf(corr)

plt.show()
```
