# Problema 1

● Se tiene una hoja metálica de una turbina como la que se muestra a la izquierda:
turbine_blade_black_dot.png

● Ésta contiene una impureza la cual se desea detectar. Utiliza el método de detección de puntos donde el
kernel es una matriz de 3x3. Para mejorar la estimación modifica el kernel en tamaño o en coeficientes de
ponderación.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


img = cv2.imread('turbine_blade_black_dot.png')

kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])


kernelr = cv2.filter2D(img, -1, kernel)


kernellaplaciano = np.array([ [0, 1, 0],
                              [1,-4, 1],
                              [0, 1, 0]])

kernelgrande = np.array([ [-1, 1, -1, -1, -1],
                          [-1, 4, 4, 4, -1],
                          [-1, 4, 8, 4 ,-1],
                          [-1, 4, 4, 4, -1],
                          [-1,-1,-1,-1,-1]])


kernelmodificado = cv2.filter2D(img, -1, kernellaplaciano)


plt.figure(figsize=(20, 10))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(kernelr,cmap='gray')
plt.title("Kernel R")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(kernelmodificado,cmap='gray')
plt.title("Kernel Modificado")
plt.axis("off")
```

#Problema 2

● Se tiene una hoja metálica de una turbina como la que se muestra a la izquierda:
turbine_blade_black_dot.png

● Ésta contiene una impureza la cual se desea detectar. Utiliza el método de detección de puntos donde el
kernel es una matriz de 3x3. Para mejorar la estimación modifica el kernel en tamaño o en coeficientes de
ponderación.

```python
img = cv2.imread('wirebond_mask.png')

kernelHV = np.array([[-1, -1, -1],
                   [2,2, 2],
                   [-1, -1, -1]])

kernel45 = np.array([[-1, -1, 2],
                   [-1, 2, -1],
                   [2, -1, -1]])

kernel452 = np.array([[2, -1, -1],
                   [-1, 2, -1],
                   [-1, -1, 2]])


horizontal = cv2.filter2D(img, -1, kernelHV)
vertical = cv2.filter2D(img, -1, np.transpose(kernelHV))
grados = cv2.filter2D(img, -1, kernel45)
grados2= cv2.filter2D(img, -1, kernel452)


plt.figure(figsize=(20, 10))

plt.subplot(1, 5, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 5, 2)
plt.imshow(horizontal, cmap='gray')
plt.title("Horizontal")
plt.axis("off")


plt.subplot(1, 5, 3)
plt.imshow(vertical, cmap='gray')
plt.title("Vertical")
plt.axis("off")


plt.subplot(1, 5, 4)
plt.imshow(grados, cmap='gray')
plt.title("45 grados")
plt.axis("off")

plt.subplot(1, 5, 5)
plt.imshow(grados2, cmap='gray')
plt.title("-45 grados")
plt.axis("off")
```

#Problema 3

Realiza un programa en python que a partir de Prewitt en la imagen original building.png calcule:

● El componente |Gx
| el gradiente en dirección horizontal

● El componente |Gy
| el gradiente en dirección vertical

● La imagen total de gradiente |Gx
| + |Gy
|


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


img = cv2.imread('building.png')

kernelgx = np.array([[-1, -1, -1],
                    [0,  0, 0],
                    [1, 1, 1]])


GX = cv2.filter2D(img, -1, kernelgx)
GY = cv2.filter2D(img, -1, np.transpose(kernelgx))


gradiente = cv2.add(GX,GY)


plt.figure(figsize=(20, 10))


plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(GX, cmap='gray')
plt.title("Gx")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(GY, cmap='gray')
plt.title("Gy")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(gradiente, cmap='gray')
plt.title("Gx + Gy")
plt.axis("off")
```

#Problema 4
Realiza un programa en python que a partir de Sobel en la imagen original building.png calcule:

● El componente |Gx
| el gradiente en dirección horizontal

● El componente |Gy
| el gradiente en dirección vertical

● La imagen total de gradiente |Gx
| + |Gy
|

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


img = cv2.imread('building.png')

kernelgx = np.array([[-1, -2, -1],
                    [0,  0, 0],
                    [1, 2, 1]])


GX = cv2.filter2D(img, -1, kernelgx)
GY = cv2.filter2D(img, -1, np.transpose(kernelgx))


gradiente = cv2.add(GX,GY)


plt.figure(figsize=(20, 10))


plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(GX, cmap='gray')
plt.title("Gx")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(GY, cmap='gray')
plt.title("Gy")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(gradiente, cmap='gray')
plt.title("Gx + Gy")
plt.axis("off")
```

```python

kernelpre = np.array([[-1, -1, -1],
                    [0,  0, 0],
                    [1, 1, 1]])


GXp = cv2.filter2D(img, -1, kernelpre)
GYp = cv2.filter2D(img, -1, np.transpose(kernelpre))


gradientep = cv2.add(GXp,GYp)

plt.figure(figsize=(20, 10))


plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(GXp,cv2.COLOR_BGR2RGB))
plt.title("Gx")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(GYp,cv2.COLOR_BGR2RGB))
plt.title("Gy")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(gradientep,cv2.COLOR_BGR2RGB))
plt.title("Gx + Gy")
plt.axis("off")
```
