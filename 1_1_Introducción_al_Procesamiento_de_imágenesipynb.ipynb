# Ejercicio 4
● Realiza un programa en python que permita cambiar de resolución la imagen cactus.png

(a) Convertir a imagen de 8 bits

(b) Convertir a imagen de 4 bits

(c) Convertir a imagen de 2 bits

● Imprimir en pantalla cuántos niveles de intensidad discretos posibles se obtienen en cada inciso.

Recomendación: utilizar opencv con sólo funciones de numpy, como power(), around() y astype()

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
cactus = cv2.imread('cactus.png',1)

print("Imagen de 2 bits")
k2=2
L2=2**k2
xq2c = cactus.astype(np.float32)
xq2c = np.round(cactus*(L2/255))*(255/L2)
xq2c = np.uint8(xq2c)
cv2_imshow(xq2c)

print("Imagen de 4 bits")
k4=4
L4=2**k4
xq4 = cactus.astype(np.float32)
xq4 = np.round(cactus*(L4/255))*(255/L4)
xq4 = np.uint8(xq4)
cv2_imshow(xq4)

print("Imagen de 8 bits")
k8=8
L8=2**k8
xq8 = cactus.astype(np.float32)
xq8 = np.round(cactus*(L8/255))*(255/L8)
xq8 = np.uint8(xq8)
cv2_imshow(xq8)
```

# Ejercicio 6
● De la imagen sunflowers.png realizar un programa en
python que permita convertir la imagen original a escala
de grises, excepto un color en específico, por ejemplo el
color amarillo

● For por renglon

  ○ For por columna

  ● If amarillo

  Conservar el pixel igual
  
● Else

Realizar escala de gris para ese pixel, (R+G+B)/3

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

amarillobajo = np.array([12, 100, 100], dtype=np.uint8)
amarilloalto = np.array([35, 255, 255], dtype=np.uint8)


sunflower = cv2.imread('sunflower.png')

SunfloweGris = cv2.cvtColor(sunflower, cv2.COLOR_BGR2GRAY)
SunfloweGris = cv2.cvtColor(SunfloweGris, cv2.COLOR_GRAY2BGR)
SunflowerHSV = cv2.cvtColor(sunflower, cv2.COLOR_BGR2HSV)

MascaraAmarilla = cv2.inRange(SunflowerHSV, amarillobajo, amarilloalto )
MascaraAmarilla = cv2.medianBlur(MascaraAmarilla,7)

MascaraAmarillavis = cv2.bitwise_and(sunflower,sunflower, mask=MascaraAmarilla)

MascaraInv = cv2.bitwise_not(MascaraAmarilla)
FondoGris = cv2.bitwise_and(SunfloweGris,SunfloweGris,mask=MascaraInv)

sunflowerGrisAmarillo = cv2.add(FondoGris,MascaraAmarillavis)

cv2_imshow(sunflowerGrisAmarillo)
```
